##  Step 1: Import Libraries & Load Dataset

```python
# Importing required libraries for data handling, visualization, and loading the dataset
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
# Loading the already built-in breast cancer dataset from sklearn.
cancer = load_breast_cancer()
```

## Step 2: Exploratory Data Analysis (EDA)

```python
#Convert to DataFrame
```

```python
# Converting the dataset into a pandas DataFrame for exploration.
df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
```

```python
# Adding the target column (0 = Malignant, 1 = Benign) to the DataFrame.
df['target'] = cancer.target
```

```python
# View Basic Information
```

```python
# viewing the dataset shape, first few rows, and data types to understand the structure.
print(df.head())       
print(df.info())      
print(df.describe())   
```

```python
#Plot Class Distribution
```

```python
# Ploting the number of malignant (0) vs benign (1) cases to check data balance.
sns.countplot(x='target', data=df)
plt.title("Class Distribution (0 = Malignant, 1 = Benign)")
plt.show()
```

```python
#Correlation Heatmap
```

```python
# Visualizing the feature correlations to understand how they relate to each other and the targets.
plt.figure(figsize=(18,10))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Step 3: Data Preprocessing

```python
# Checking & Handling the Missing Values in the set
```

```python
# Checking for missing values
df.isnull().sum()

# If it contain we will fill with the mean value
df.fillna(df.mean(), inplace=True)  

#found no missing value
```

```python
#Encoding the Categorical Labels (e.g., 'M' and 'B')
```

```python
# Converting the diagnosis column to binary column
df['target'] = df['target'].map({'M': 1, 'B': 0})
```

```python
#Apply StandardScaler
```

```python
# doing this to normalize all features so that they have the same scale, improving model performance and training stability
from sklearn.preprocessing import StandardScaler

# Separating the features and labels here
X = df.drop('target', axis=1)
y = df['target']

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Train-Test Split
```

```python
#doing thisto divide the data into training and testing sets so the model can learn from one part and be tested on unseen data to check its performance.
from sklearn.model_selection import train_test_split

# Spliting the data into 80%  for training and  20% testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```
